### Quantifying the motion outliers in every run and subject
#### Threshold was .9

```python
import pandas as pd
from glob import glob
import sys
import seaborn as sns
import csv
from IPython.display import Image
import matplotlib.pyplot as plt
import json
import os
%pylab
%matplotlib inline
```

### Construct a dataframe that includes the number of motion outlier timepoints

```python
run1 = pd.DataFrame()
run2 = pd.DataFrame()
run3 = pd.DataFrame()
run4 = pd.DataFrame()
run5 = pd.DataFrame()
run6 = pd.DataFrame()

path = '/misc/data19/schonehr/ProControl/data/DONE/'
subjects = [name for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))]

for subj in subjects:
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_1_bold.txt') as f:
            line = f.readline()
            run1_value = len(line.split())
            run1 = run1.append([[subj,run1_value]],sort=False)
    except OSError:
        print(subj + ': run1 file not found')
        run1 = run1.append([[subj,]],sort=False)      
        pass
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_2_bold.txt') as f:
            line = f.readline()
            run2_value = len(line.split())
            run2 = run2.append([[subj,run2_value]],sort=False)
    except OSError:
        print(subj + ': run2 file not found')
        run2 = run2.append([[subj,]],sort=False)      
        pass
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_3_bold.txt') as f:
            line = f.readline()
            run3_value = len(line.split())
            run3 = run3.append([[subj,run3_value]],sort=False)
    except OSError:
        print(subj + ': run3 file not found')
        run3 = run3.append([[subj,]],sort=False)      
        pass
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_4_bold.txt') as f:
            line = f.readline()
            run4_value = len(line.split())
            run4 = run4.append([[subj,run4_value]],sort=False)
    except OSError:
        print(subj + ': run4 file not found')
        run4 = run4.append([[subj,]],sort=False)      
        pass
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_5_bold.txt') as f:
            line = f.readline()
            run5_value = len(line.split())
            run5 = run5.append([[subj,run5_value]],sort=False)
    except OSError:
        print(subj + ': run5 file not found')
        run5 = run5.append([[subj,]],sort=False)      
        pass
    try:
        with open('/misc/data19/schonehr/ProControl/data/DONE/' + subj + '/pre/func/motion_assess/confound_6_bold.txt') as f:
            line = f.readline()
            run6_value = len(line.split())
            run6 = run6.append([[subj,run6_value]],sort=False)      
    except OSError:
        print(subj + ': run6 file not found')
        run6 = run6.append([[subj,]],sort=False)      
        pass

run1.columns = ['subj','run1']
run1 = run1.sort_values(by =['subj'])
run2.columns = ['subj','run2']
run2 = run2.sort_values(by =['subj'])
run3.columns = ['subj','run3']
run3 = run3.sort_values(by =['subj'])
run4.columns = ['subj','run4']
run4 = run4.sort_values(by =['subj'])
run5.columns = ['subj','run5']
run5 = run5.sort_values(by =['subj'])
run6.columns = ['subj','run6']
run6 = run6.sort_values(by =['subj'])
runs= ['run1','run2','run3','run4','run5','run6']

df = pd.concat([run1,run2,run3,run4,run5,run6], axis=1)
df = df.loc[:, ~df.columns.duplicated()]
df['group']=df['subj'].str.slice(start=4,stop=5)
df['group'] = df['group'].str.replace('c','0')
df['group'] = df['group'].str.replace('a','2')
df['group'] = df['group'].str.replace('b','1')
df['sum'] = sum(df[runs].values,axis=1)
df['percentage'] = df['sum']/1968 * 100
print(df) 
df.to_csv('/misc/data19/schonehr/ProControl/analysis/mri_preprocessing/motion_outliers/all_motion_outlier.csv')
```

```python
df.plot.scatter(x='subj',y='sum')
plt.title('Sum of all motion outliers')
df.plot.scatter(x='subj',y='percentage')
plt.title('Percentage of motion outliers of total volumes')
```

### Motion outliers for each run for individual subjects

```python
fig, axes = plt.subplots(nrows=1,ncols=6,figsize=(25,6))
run1 = df.plot.scatter(x='subj',y='run1',colormap='viridis',ax = axes[0],subplots=True)
run2 = df.plot.scatter(x='subj',y='run2',colormap='viridis',ax = axes[1],subplots=True)
run3 = df.plot.scatter(x='subj',y='run3',colormap='viridis',ax = axes[2],subplots=True) 
run4 = df.plot.scatter(x='subj',y='run4',colormap='viridis',ax = axes[3],subplots=True) 
run5 = df.plot.scatter(x='subj',y='run5',colormap='viridis',ax = axes[4],subplots=True) 
run6 = df.plot.scatter(x='subj',y='run6',colormap='viridis',ax = axes[5],subplots=True) 
plt.show()
```

### Group comparison of motion outliers for each run

```python
runs= ['run1','run2','run3','run4','run5','run6']
df.groupby('group')[runs].mean().T.plot.bar(yerr=df.groupby('group')[runs].sem().T,figsize=(12,8),fontsize=12,width=.9)
plt.ylabel('Average number of motion outliers',fontsize=12)
plt.xlabel('Runs',fontweight='bold',fontsize=16)
plt.xticks([0,1,2,3,4,5],['run1','run2','run3','run4','run5','run6'],rotation='horizontal', fontsize=10)
```
